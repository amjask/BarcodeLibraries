This notebook describes the process for generating a 4^12 combinatorial library from a list of verified imagable 20bp sequences. We call each of these 20bp sequences a "barcode". The library is made by fusing 2 oligo pools each containing 4^6 combinations. The original intent was to have 4^5 oligos in each pool, keeping out 1 barcode per pool to be used as a parity bit for error correction. However the recombinations during oligo synthesis and/or cloning resulted in having many if not all possible combinations of each oligo pool in the final library.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import/create the sequences

# barcode sequences from our screening
ver_BCs = pd.read_csv('verified_barcodes_200305.csv')
ver_BCs.sort_values('quality', inplace=True, ignore_index=True)
bc_list = np.asarray(ver_BCs.sequence).reshape([4,12])

# sequences used in cloning
spacers = np.array(['AACCT','ACGTC','ACTGC','CTAAC','CATTG','TTGCC',
                    'CGATG','CATAC','CCAGT','TCTGT','GTGTC']);
primers = np.array([['CGTAAACGGCCACAAGTTCG','CGCACGACTTCTTCAAGTCC'],
                    ['AAGGGCATCGACTTCAAGCG','CGCAGAAGAACGGCATCAAG']])
BpiI_5 = 'GAAGACagATGG'
BpiI_3 = 'CTCCctGTCTTC'
```

```python
# generate the codes
codes = []

for pool in range(2):
    code_list = []
    for i in range(4**5):
        base = np.base_repr(i, base=4).zfill(5)
        base_list = list(map(int,(list(base))))
        parity = sum(base_list)%4
        base_list.insert(0,parity)
        code_list.append(base_list)
        
    codes.append(code_list)
    
codes = np.array(codes)
```

```python
# create the oligo sequences
oligos = []
# first pool
for i in range(4**5):
    o = primers[0][0]
    for j in range(6):
        o = o + bc_list[codes[0][i][j],j] + spacers[j]
    o = o[:-5] + primers[0][1]
    oligos.append(o)

# second pool
for i in range(4**5):
    o = primers[1][0]
    for j in range(6):
        o = o + bc_list[codes[1][i][j],j+6] + spacers[j+5]
    o = o[:-5] + primers[1][1]
    oligos.append(o)
```

```python
# create all the combinations
combinations = []
for i in range(4**5):
    for j in range(4**5):
        combinations.append(oligos[i] + oligos[j+4**5])
        
# test them for BsaI, BpiI, Esp3I cut sites, as well as 5bp long homopolymers
BsaI = ['GGTCTC','GAGACC']
BpiI = ['GAAGAC','GTCTTC']
Esp3I = ['GAGACG','CGTCTC']
homopolymer = ['AAAAA','TTTTT','CCCCC','GGGGG']

bsaI_ = []
esp3I_ = []
homo_ = []
bpiI_ = []

for i in range(len(combinations)):
    query = combinations[i].upper()
    bsaI_.append(BsaI[0] in query or BsaI[1] in query)
    esp3I_.append(Esp3I[0] in query or Esp3I[1] in query)
    homo_.append('AAAAA' in query or 'TTTTT' in query or 'CCCCC' in query or 'GGGGG' in query)
    bpiI_.append(BpiI[0] in query or BpiI[1] in query)

df = pd.DataFrame(list(zip(bsaI_, esp3I_, homo_, bpiI_, combinations)), 
               columns =['BsaI', 'Esp3I', 'Homopolymer', 'BpiI', 'Combination']) 


for i in ver_BCs['sequence']:
    col = ver_BCs.loc[ver_BCs['sequence'] == i]['name'].values[0]
    val = []
    for j in range(len(combinations)):
            query = combinations[j].upper()
            val.append(query.count(i.upper()))
    
    df[col] = val
  

print('Does not include any BsaI site: {}'.format(sum(df['BsaI']) == 0))
print('Does not include any Esp3I site: {}'.format(sum(df['Esp3I']) == 0))
print('Does not include any BpiI site: {}'.format(sum(df['BpiI']) == 0))
print('Does not include any homopolyer 5bp or longer: {}'.format(sum(df['Homopolymer']) == 0))

bc_check = []
for i in range(5,53):
    bc_check.append(sum(df.iloc[:,i]) == len(df)/4)

print('Each barcode is used in 1/4 of the combinations: {}'.format(sum(bc_check) == 48))
```

```python
# visualize the oligos to make sure everything is correct
bc_in_oligo = []
for i in ver_BCs['sequence']:
    col = ver_BCs.loc[ver_BCs['sequence'] == i]['name'].values[0]
    val = []
    for j in range(len(oligos)):
            query = oligos[j].upper()
            val.append(query.count(i.upper()))
    bc_in_oligo.append(val)

# fig = imagesc.seaborn(np.transpose(np.asarray(bc_in_oligo)))
plt.figure(figsize=(10,2), dpi=150)
plt.imshow(np.transpose(np.asarray(bc_in_oligo)), aspect='auto')
plt.yticks(np.arange(0,2048,256))
plt.xticks(np.arange(48), labels=np.arange(1,49,1), rotation='vertical')
plt.xlabel('barcode #')
plt.ylabel('combination #')
plt.show()
```

```python
# export sequences in Twist format 
name = []
sequence = []
for pool in range(2):
    for oligo in range(4**5):
        name.append('lib_1e6_pool{p:d}_bc'.format(p = pool + 1) + np.base_repr(oligo, base=4).zfill(5))
        sequence.append(oligos[oligo + pool*4**5])
        
for pool in range(2):
    for oligo in range(4**5):
        name.append('lib_1e6_pool{p:d}_bc_'.format(p = pool + 1) + np.base_repr(oligo, base=4).zfill(5))
        sequence.append(oligos[oligo + pool*4**5])
        
T = pd.DataFrame(list(zip(name,sequence)), columns =['name', 'sequence'])
T.to_csv('lib_1e6_oligos.txt', index=False)
```
